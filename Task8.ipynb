Calculate the mean, median, and mode of the sepal lengths in the Iris dataset

```python
import seaborn as sns
import pandas as pd
from scipy import stats
iris = sns.load_dataset('iris')
sepal_l = iris['sepal_length']
mean_sl = sepal_l.mean()
median_sl = sepal_l.median()
mode_sl = sepal_l.mode()
print(f"Mean sepal length: {mean_sl:.2f}")
print(f"Median sepal length: {median_sl}")
print(f"Mode sepal length: {mode_sl}")
```

Calculate the variance and standard deviation of the petal widths in the Iris dataset.

```python
petal_widths = iris['petal_width']
import numpy as np
variance_pw = np.var(petal_widths)
std_dev_pw = np.std(petal_widths)
print(f"Variance of petal widths: {variance_pw:.2f}")
print(f"Standard deviation of petal widths: {std_dev_pw:.2f}")
```

Create a summary table that includes the mean, median, variance, and standard deviation for all numerical features in the dataset.

```python
numerical_features = iris.select_dtypes(include=[np.number])
summary_table = numerical_features.aggregate(['mean', 'median', 'var', 'std']).transpose()
summary_table.columns = ['Mean', 'Median', 'Variance', 'Std. Deviation']
print("Summary Table for Numerical Features in Iris Dataset:")
print(summary_table)
```

Define a random variable for the sepal length and calculate the probability distribution of sepal lengths.


```python
sl=sepal_l
sns.histplot(sl, kde=True, stat='density')
import matplotlib.pyplot as plt
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Probability Density')
plt.title('Probability Distribution of Sepal Lengths')
plt.show()
```

Calculate the cumulative distribution function (CDF) for the petal lengths and plot it.

```python
petal_l = iris['petal_length']
sorted_data = np.sort(petal_l)
yvals = np.arange(len(sorted_data)) / float(len(sorted_data))
plt.figure(figsize=(8, 6))
plt.plot(sorted_data, yvals, marker='.', linestyle='none')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution Function (CDF) of Petal Lengths')
plt.show()
```

Calculate and plot the probability density function (PDF) for sepal width.

```python
sepal_w = iris['sepal_width']
plt.figure(figsize=(8, 6))
sns.kdeplot(sepal_w, shade=True)
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Probability Density')
plt.title('Probability Density Function (PDF) of Sepal Widths')
plt.grid(True)
plt.show()
```

Determine the probability of a randomly selected iris flower having a petal length greater than a given value.

```python
sorted_pl = np.sort(petal_l)
cumulative_prob = np.arange(len(sorted_pl)) / len(sorted_pl)
def probability_greater_than(given_value):
    count_greater = np.sum(sorted_pl > given_value)
    return count_greater / len(sorted_pl)
given_value = 5.0
prob_greater_than_given_value = probability_greater_than(given_value)
print(f"Probability of a randomly selected iris flower having a petal length greater than {given_value} cm: {prob_greater_than_given_value:.4f}")
```

Perform a hypothesis test to determine if there is a significant difference in the mean petal length between two species of iris flowers.

```python
setosa_petal_lengths = iris.loc[iris['species'] == 'setosa', 'petal_length']
versicolor_petal_lengths = iris.loc[iris['species'] == 'versicolor', 'petal_length']
t_statistic, p_value = stats.ttest_ind(setosa_petal_lengths, versicolor_petal_lengths)
alpha = 0.05
print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
if p_value < alpha:
    print("Reject null hypothesis: There is a significant difference in mean petal length between the two species.")
else:
    print("Fail to reject null hypothesis: There is no significant difference in mean petal length between the two species.")
```

Calculate and interpret the covariance and correlation between sepal length and sepal width.

```python
sepal_l = iris['sepal_length']
sepal_w = iris['sepal_width']
covariance = sepal_l.cov(sepal_w)
correlation = sepal_l.corr(sepal_w)
print(f"Covariance between sepal length and sepal width: {covariance:.4f}")
print(f"Correlation between sepal length and sepal width: {correlation:.4f}")
```
