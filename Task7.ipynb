 Create a simple line plot using Matplotlib to display the trend of a list of values over time.

```python
import pandas as pd
import matplotlib.pyplot as plt
titanic_data = pd.read_csv('/content/titanic.csv')
titanic_data.head()
data = titanic_data[['PassengerId', 'Age']]
data['Age'].fillna(data['Age'].mean(), inplace=True)
plt.figure(figsize=(10, 6))
plt.plot(data['PassengerId'], data['Age'], label='Age', color='blue', marker='o')
plt.title('Trend of Passenger Ages over PassengerId')
plt.xlabel('PassengerId')
plt.ylabel('Age')
plt.legend()
plt.show()
```

Plot a bar chart using Matplotlib to show the frequency of different categories in a dataset.

```python
pclass_counts = titanic_data['Pclass'].value_counts()
plt.figure(figsize=(8, 6))
pclass_counts.plot(kind='bar', color=['blue', 'green', 'red'])
plt.title('Frequency of Different Passenger Classes')
plt.xlabel('Passenger Class')
plt.ylabel('Frequency')
plt.show()
```

Create a scatter plot using Matplotlib to visualize the relationship between two variables in a dataset.

```python
age_fare_data = titanic_data[['Age', 'Fare']].dropna()
plt.figure(figsize=(10, 6))
plt.scatter(age_fare_data['Age'], age_fare_data['Fare'], alpha=0.5, color='blue')
plt.title('Scatter Plot of Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

 Load a dataset using Seaborn's built-in dataset functions and create a pairplot to visualize the relationships between all pairs of features.

```python
import seaborn as sns
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species')
plt.show()
```

Create a box plot using Seaborn to show the distribution of values for different categories in a dataset.

```python
titanic = sns.load_dataset('/content/titanic')
plt.figure(figsize=(10, 6))
sns.boxplot(x='pclass', y='age', data=titanic)
plt.title('Box Plot of Age Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```

Plot a heatmap using Seaborn to visualize the correlation matrix of a dataset.

```python
numeric_cols = titanic.select_dtypes(include='number')
correlation_matrix = numeric_cols.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

Use Matplotlib to create a subplot grid that displays multiple charts in a single figure.


```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
axes[0, 0].hist(titanic['age'].dropna(), bins=30, color='blue', edgecolor='black')
axes[0, 0].set_title('Histogram of Age')
axes[0, 0].set_xlabel('Age')
axes[0, 0].set_ylabel('Frequency')
sns.boxplot(x='pclass', y='age', data=titanic, ax=axes[0, 1])
axes[0, 1].set_title('Box Plot of Age by Passenger Class')
axes[0, 1].set_xlabel('Passenger Class')
axes[0, 1].set_ylabel('Age')
sns.countplot(x='pclass', data=titanic, ax=axes[1, 0])
axes[1, 0].set_title('Passenger Class Distribution')
axes[1, 0].set_xlabel('Passenger Class')
axes[1, 0].set_ylabel('Count')
axes[1, 1].scatter(titanic['age'], titanic['fare'], alpha=0.5, color='green')
axes[1, 1].set_title('Scatter Plot of Age vs Fare')
axes[1, 1].set_xlabel('Age')
axes[1, 1].set_ylabel('Fare')
plt.tight_layout()
plt.show()
```

 Customize the appearance of a Seaborn plot by changing the color palette, adding titles, and modifying axis labels.

```python
custom_palette = sns.color_palette("hls", 8)
plt.figure(figsize=(5, 3))
sns.boxplot(x='pclass', y='age', data=titanic, palette=custom_palette)
plt.title('Box Plot of Age Distribution(y) by Passenger Class(x)', fontsize=8)
plt.xlabel('Passenger Class', fontsize=7)
plt.ylabel('Age', fontsize=7)
plt.xticks(fontsize=5)
plt.yticks(fontsize=5)
plt.show()
```

Create a violin plot using Seaborn to visualize the distribution of a dataset across different categories.

```python
sns.violinplot(x='pclass', y='age', data=titanic)
plt.title('Violin Plot of Age Across Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```

Combine Matplotlib and Seaborn to create a complex visualization, such as overlaying a KDE plot on a histogram.

```python
sns.histplot(data=titanic, x="age", bins=30, color='blue')
sns.kdeplot(data=titanic, x="age", color='red', shade=True)
plt.title('Distribution of Age on the Titanic')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
