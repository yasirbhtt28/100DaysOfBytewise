Handling Outliers in the Boston Housing Dataset

```python
import numpy as np
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
boston_df = pd.DataFrame(data, columns=[
    'CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT'
])
boston_df['MEDV'] = target
boston_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(20, 15))
for i, col in enumerate(boston_df.columns):
    plt.subplot(5, 3, i + 1)
    sns.boxplot(y=boston_df[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
from scipy import stats
z_scores = stats.zscore(boston_df)
z_scores_df = pd.DataFrame(z_scores, columns=boston_df.columns)
z_score_outliers = (z_scores_df > 3) | (z_scores_df < -3)
z_score_outliers.sum()
```

```python
Q1 = boston_df.quantile(0.25)
Q3 = boston_df.quantile(0.75)
IQR = Q3 - Q1
iqr_outliers = (boston_df < (Q1 - 1.5 * IQR)) | (boston_df > (Q3 + 1.5 * IQR))
iqr_outliers.sum()
```

```python
boston_df_no_outliers = boston_df[~(iqr_outliers.any(axis=1))]
original_shape = boston_df.shape
cleaned_shape = boston_df_no_outliers.shape
original_shape, cleaned_shape
```

```python
z_score_outliers_sum = z_score_outliers.sum()
iqr_outliers_sum = iqr_outliers.sum()
z_score_outliers_sum, iqr_outliers_sum, original_shape, cleaned_shape
```

Data Imputation in the Retail Sales Dataset

```python
retail_df = pd.read_csv("/content/retail_sales_dataset.csv")
retail_df.head(), retail_df.isnull().sum()
```

```python
import numpy as np
np.random.seed()
missing_data_df = retail_df.copy()
missing_data_df.loc[np.random.choice(missing_data_df.index, size=5), 'Quantity'] = np.nan
missing_data_df.loc[np.random.choice(missing_data_df.index, size=5), 'Price per Unit'] = np.nan
missing_data_df.loc[np.random.choice(missing_data_df.index, size=5), 'Total Amount'] = np.nan
missing_data_df.head(10), missing_data_df.isnull().sum()
```

```python
from sklearn.impute import KNNImputer
knn_imputer = KNNImputer(n_neighbors=2)
knn_imputed = knn_imputer.fit_transform(missing_data_df.select_dtypes(include=[np.number]))
knn_imputed_df = pd.DataFrame(knn_imputed, columns=missing_data_df.select_dtypes(include=[np.number]).columns)
knn_imputed_df = pd.concat([missing_data_df.select_dtypes(exclude=[np.number]).reset_index(drop=True), knn_imputed_df.reset_index(drop=True)], axis=1)
knn_imputed_df.head(10)
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
mice_imputer = IterativeImputer(max_iter=10, random_state=0)
mice_imputed = mice_imputer.fit_transform(missing_data_df.select_dtypes(include=[np.number]))
mice_imputed_df = pd.DataFrame(mice_imputed, columns=missing_data_df.select_dtypes(include=[np.number]).columns)
mice_imputed_df = pd.concat([missing_data_df.select_dtypes(exclude=[np.number]).reset_index(drop=True), mice_imputed_df.reset_index(drop=True)], axis=1)
mice_imputed_df.head(10)
```

```python
print("Original Dataset with Missing Values:")
print(missing_data_df.head(10))
print("\nKNN Imputed Dataset:")
print(knn_imputed_df.head(10))
print("\nMICE Imputed Dataset:")
print(mice_imputed_df.head(10))
```

Feature Engineering in the Heart Disease Dataset


```python
heart_df = pd.read_csv("/content/heart_disease.csv")
heart_df.head()
```

```python
age_bins = [0, 30, 40, 50, 60, 70, 80, 100]
age_labels = ['<30', '30-39', '40-49', '50-59', '60-69', '70-79', '80+']
heart_df['age_group'] = pd.cut(heart_df['age'], bins=age_bins, labels=age_labels, right=False)
heart_df['age_group'].value_counts()
```

```python
heart_rate_bins = [0, 90, 120, 150, 200]
heart_rate_labels = ['Low', 'Normal', 'Elevated', 'High']
heart_df['heart_rate_level'] = pd.cut(heart_df['thalach'], bins=heart_rate_bins, labels=heart_rate_labels, right=False)
heart_df['heart_rate_level'].value_counts()
```

Transforming Variables in the Bike Sharing Dataset

```python
bike_df = pd.read_csv("/content/day.csv")
bike_df.head()
```

```python
from scipy.stats import skew
numeric_cols = bike_df.select_dtypes(include=[np.number]).columns
skewness = bike_df[numeric_cols].apply(skew)
skewed_vars = skewness[abs(skewness) > 0.5].index
skewness, skewed_vars
```

```python
log_transformed_df = bike_df.copy()
for col in skewed_vars:
    log_transformed_df[col] = np.log1p(log_transformed_df[col])
log_transformed_df[skewed_vars].head()
```

```python
sqrt_transformed_df = bike_df.copy()
for col in skewed_vars:
    sqrt_transformed_df[col] = np.sqrt(sqrt_transformed_df[col])
sqrt_transformed_df[skewed_vars].head()
```

```python
from scipy.stats import boxcox
boxcox_transformed_df = bike_df.copy()
for col in skewed_vars:
    if (boxcox_transformed_df[col] > 0).all():
        boxcox_transformed_df[col], _ = boxcox(boxcox_transformed_df[col])
boxcox_transformed_df[skewed_vars].head()
```

```python
log_transformed_head = log_transformed_df[skewed_vars].head()
sqrt_transformed_head = sqrt_transformed_df[skewed_vars].head()
boxcox_transformed_head = boxcox_transformed_df[skewed_vars].head()
skewness, skewed_vars, log_transformed_head, sqrt_transformed_head, boxcox_transformed_head
```

Feature Selection in the Diabetes Dataset

```python
import pandas as pd
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
diabetes_df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
diabetes_df['target'] = diabetes.target
diabetes_df.head()
```

```python
correlation_matrix = diabetes_df.corr()
correlation_with_target = correlation_matrix['target'].sort_values(ascending=False)
correlation_with_target
```

```python
from sklearn.feature_selection import mutual_info_regression
X = diabetes_df.drop('target', axis=1)
y = diabetes_df['target']
mutual_info = mutual_info_regression(X, y)
mutual_info_df = pd.DataFrame(mutual_info, index=X.columns, columns=['Mutual Information'])
mutual_info_df.sort_values(by='Mutual Information', ascending=False)
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
model = LinearRegression()
rfe = RFE(model, n_features_to_select=5)
rfe.fit(X, y)
rfe_ranking = pd.DataFrame(rfe.ranking_, index=X.columns, columns=['RFE Ranking']).sort_values(by='RFE Ranking')
rfe_ranking
```

Dealing with Imbalanced Data in the Credit Card Fraud Detection Dataset

```python
url = "https://storage.googleapis.com/download.tensorflow.org/data/creditcard.csv"
credit_df = pd.read_csv(url)
credit_df.head()
```

```python
import pandas as pd
from imblearn.over_sampling import SMOTE, ADASYN
from imblearn.under_sampling import RandomUnderSampler
class_distribution = credit_df['Class'].value_counts()
print("Original class distribution:")
print(class_distribution)
X = credit_df.drop('Class', axis=1)
y = credit_df['Class']
```

```python
smote = SMOTE(random_state=42)
X_smote, y_smote = smote.fit_resample(X, y)
smote_class_distribution = pd.Series(y_smote).value_counts()
print("\nClass distribution after SMOTE:")
print(smote_class_distribution)
```

```python
adasyn = ADASYN(random_state=42)
X_adasyn, y_adasyn = adasyn.fit_resample(X, y)
adasyn_class_distribution = pd.Series(y_adasyn).value_counts()
print("\nClass distribution after ADASYN:")
print(adasyn_class_distribution)
```

```python
undersampler = RandomUnderSampler(random_state=42)
X_under, y_under = undersampler.fit_resample(X, y)
under_class_distribution = pd.Series(y_under).value_counts()
print("\nClass distribution after undersampling:")
print(under_class_distribution)
```

Combining Multiple Datasets in the Movie Lens Dataset

```python
import pandas as pd
import zipfile
with zipfile.ZipFile('movielens-20m-dataset.zip', 'r') as zip_ref:
    zip_ref.extractall('movielens-20m-dataset')
ratings = pd.read_csv('/content/movielens-20m-dataset/movielens-20m-dataset/ratings.csv')
movies = pd.read_csv('/content/movielens-20m-dataset/movielens-20m-dataset/movies.csv')
tags = pd.read_csv('/content/movielens-20m-dataset/movielens-20m-dataset/tags.csv')
links = pd.read_csv('/content/movielens-20m-dataset/movielens-20m-dataset/links.csv')
print("Ratings:")
print(ratings.head())
print("\nMovies:")
print(movies.head())
print("\nTags:")
print(tags.head())
print("\nLinks:")
print(links.head())
```

```python
print("Ratings Info:")
print(ratings.info())
print("\nMovies Info:")
print(movies.info())
print("\nTags Info:")
print(tags.info())
print("\nLinks Info:")
print(links.info())
```

```python
merged_df = ratings.merge(movies, on='movieId')
merged_df = merged_df.merge(tags, on=['movieId', 'userId'], how='left')
merged_df = merged_df.merge(links, on='movieId', how='left')
print("Merged Dataset:")
print(merged_df.head())
```

```python
missing_values = merged_df.isnull().sum()
print("Missing Values:")
print(missing_values)
```

```python
merged_df.dropna(inplace=True)
```

```python
missing_values = merged_df.isnull().sum()
print("Missing Values:")
print(missing_values)
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
label_encoder = LabelEncoder()
merged_df['genres'] = label_encoder.fit_transform(merged_df['genres'].astype(str))
scaler = StandardScaler()
merged_df[['rating', 'timestamp_x']] = scaler.fit_transform(merged_df[['rating', 'timestamp_x']])
merged_df[['rating', 'timestamp_y']] = scaler.fit_transform(merged_df[['rating', 'timestamp_y']])
print("Preprocessed Dataset:")
print(merged_df.head())
```

```python
output_file_path = '/content/movielens_preprocessed.csv'
merged_df.to_csv(output_file_path, index=False)
output_file_path
```
